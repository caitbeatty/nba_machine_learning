```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
player_data_to_load = "data/Seasons_Stats_current.csv"
champion_data_to_load = "data/finals_results_by_year.csv"
runner_up_data_to_load = "data/runnerupsdata.csv"

# Read School and Student Data File and store into Pandas DataFrames
player_data = pd.read_csv(player_data_to_load)
champion_data = pd.read_csv(champion_data_to_load)
runnerup_data = pd.read_csv(runner_up_data_to_load)


# Combine the data into a single dataset.  
#school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#school_data_complete
```

```python
#just need year and Team name

winner_summary = champion_data[{"Year","Team"}]
winner_summary
```

```python
runnerup_summary = runnerup_data[{"Year","Team"}]
runnerup_summary
```

```python
#get a merged list of all winners runner up or champion
championship_complete = pd.merge(winner_summary, runnerup_summary, how="inner", on=["Year", "Year"])
```

```python
championship_complete
```

```python
#rename 
rename_summary= championship_complete.rename(columns= {"Team_x": "Winner", "Team_y": "Runner_Up"})
rename_summary.head()
```

```python
clean_winner_df = rename_summary.drop_duplicates(subset=['Year', 'Winner', "Runner_Up"], keep='first')
```

```python
clean_winner_df.head(5)
```

```python
game_data_complete = pd.merge(clean_winner_df, player_data, how="inner", on=["Year", "Year"])
```
