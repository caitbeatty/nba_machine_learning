```python
winning_team_df = pd.merge(champion_data, player_data_df, how="inner", left_on = ["Year", "Winners"], right_on = ["Year", "name"])
winning_team_df.head()
```

```python
player_winning_df = winning_team_df[["Year", "Player"]]
player_winning_df
```

```python
winner_stats_df = player_winning_df.groupby("Player").count()
winner_stats_df.rename(columns = {"Year" : "Championship Wins"}, inplace = True)
winner_stats_df.reset_index(inplace = True)
winner_stats_df
```

```python
runnerup_team_df = pd.merge(champion_data, player_data_df, how="inner", left_on = ["Year", "Runner Up"], right_on = ["Year", "name"])
runnerup_team_df.head()
```

```python
runnerup_df = runnerup_team_df[["Year", "Player"]]
player_runnerup_df = runnerup_df[runnerup_df['Year'] > 1951]
player_runnerup_df
```

```python
runnerup_stats_df = player_runnerup_df.groupby("Player").count()
runnerup_stats_df.rename(columns = {"Year" : "RunnerUp Count"}, inplace = True)
runnerup_stats_df.reset_index(inplace = True)
runnerup_stats_df
```

```python
players_updated_df = pd.merge(winner_stats_df, player_data_df, how="right", on = "Player")
players_updated_df
```

```python
players_complete_df = players_updated_df.merge(runnerup_stats_df, how = "left", on = "Player")
players_complete_df
```
