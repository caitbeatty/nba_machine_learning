```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Create Variables for csv files
player_data_to_load = "data/backupdata/Seasons_Stats.csv"
champion_data_to_load = "data/finals_results_by_year.csv"
runner_up_data_to_load = "data/runnerupsdata.csv"
salary_data_to_load = "data/NBA Players - Advanced Season Stats.csv"

# Read data into DataFrames
player_data = pd.read_csv(player_data_to_load)
champion_data = pd.read_csv(champion_data_to_load)
runnerup_data = pd.read_csv(runner_up_data_to_load)
salary_data = pd.read_csv(salary_data_to_load)
```

```python
# Review the Championship DataFrame
champion_data
```

```python
# Remove "*" denoting player performance in a particular game from Player name column
player_data["Player"] = player_data['Player'].str.replace('*','')
```

```python
# Data prior to 1951 does not include the TRB stat, filter out any data prior to 1951
player_data_df = player_data[player_data['Year'] > 1950]
player_data_df.rename(columns = {"name": "Team"}, inplace = True)
player_data_df
```

```python
# Merge championship data with player data to find players on championship teams
winning_team_df = pd.merge(champion_data, player_data_df, how="inner",\
                           left_on = ["Year", "Winners"], right_on = ["Year", "Team"])
winning_team_df.head()
```

```python
# # Create a new DataFrame with only the Year and the name of the Player who won a championship
# player_winning_df = winning_team_df[["Year", "Player"]]
# player_winning_df
```

```python
# Create a new DataFrame with only the Year and the name of the Player who won a championship
player_winning_df = winning_team_df[["Year", "Player"]]
player_winning_df["Championship Wins"] = 1
player_winning_df
```

```python
# # Use GroupBy to count the number of times a player was on a championship team
# winner_stats_df = player_winning_df.groupby("Player").count()
# winner_stats_df.rename(columns = {"Year" : "Championship Wins"}, inplace = True)
# winner_stats_df.reset_index(inplace = True)
# winner_stats_df
```

```python
# Merge championship data with player data to find players on runner up teams
runnerup_team_df = pd.merge(champion_data, player_data_df, how="inner",\
                            left_on = ["Year", "Runner Up"], right_on = ["Year", "Team"])
runnerup_team_df.head()
```

```python
# # Create a new DataFrame with only the Year and the name of the Player
# runnerup_df = runnerup_team_df[["Year", "Player"]]
# player_runnerup_df = runnerup_df[runnerup_df['Year'] > 1951]
# player_runnerup_df
```

```python
# Create a new DataFrame with only the Year and the name of the Player
runnerup_df = runnerup_team_df[["Year", "Player"]]
player_runnerup_df = runnerup_df[runnerup_df['Year'] > 1951]
player_runnerup_df["RunnerUp Count"] = 1
player_runnerup_df
```

```python
# # Use GroupBy to count the number of times a player was on a Runnerup Team
# runnerup_stats_df = player_runnerup_df.groupby("Player").count()
# runnerup_stats_df.rename(columns = {"Year" : "RunnerUp Count"}, inplace = True)
# runnerup_stats_df.reset_index(inplace = True)
# runnerup_stats_df
```

```python
# # Merge the count of championships won by players with the players DataFrame
# players_updated_df = pd.merge(winner_stats_df, player_data_df, how="right", on = "Player")
# players_updated_df
```

```python
# Merge the count of championships won by players with the players DataFrame
players_updated_df = pd.merge(player_winning_df, player_data_df, how="right", on = ["Player", "Year"])
players_updated_df
```

```python
# Merge the count of times players were in runner up teams with the updated player DataFrame
players_complete_df = players_updated_df.merge(player_runnerup_df, how = "left", on = ["Player", "Year"])
players_complete_df
```

```python
# Use the count function to check for blanks
players_complete_df.count()
```

```python
# Create a new DataFrame with select columns that demonstrate a player's performance
# Replace NaN with 0
final_df = players_complete_df[["Year", "Player", "Pos", 'Age', "Team",\
                                'Championship Wins', 'RunnerUp Count', 'TRB', 'AST',\
                                'STL', 'BLK', 'TOV', 'PF', 'PER','FG%', '3P%', 'FTA','FT%', 'PTS']]
```

```python
final_df.loc[final_df['Championship Wins'] > 0, 'Championship Appearance Y/N'] = 1
final_df.loc[final_df['RunnerUp Count'] > 0, 'Championship Appearance Y/N'] = 1
final_df = final_df.fillna(0)
final_df
# test = np.where((final_df['Championship Wins']) != 0| to_numberic(final_df['RunnerUp Count']) != 0, '1', '0')
# test
# df.loc[df['set_of_numbers'] <= 4, 'equal_or_lower_than_4?'] = 'True' 
# df.loc[df['set_of_numbers'] > 4, 'equal_or_lower_than_4?'] = 'False'
```

```python
# Save DataFrame to csv
final_df.to_csv("data/final_dataset.csv", index = False)
```

### Salary Data

```python
salary_data["truesalary"] = salary_data["truesalary"].str.replace('$','')
salary_data["truesalary"] = salary_data["truesalary"].str.replace(',','')
salary_data_df = salary_data[salary_data['year'] != 2016]
salary_data_df
```

```python
salary_final_df = salary_data_df[["year", "player", "truesalary"]]
salary_final_df.dropna(inplace = True)
salary_final_df.truesalary = pd.to_numeric(salary_final_df.truesalary, errors='coerce')
salary_final_df
```

```python
player_salary_df = salary_final_df.merge(final_df, how = "inner", left_on = ["year", "player"], right_on = ["Year", "Player"])
player_salary_df.drop(columns = ["year", "player"], inplace = True)
player_salary_df.rename(columns = {"truesalary" : "Salary"}, inplace = True)
player_salary_df
```

```python
# Save DataFrame to csv
player_salary_df.to_csv("data/Salary_Stats_clean.csv", index = False)
```

## Filter for only one salary record per player

```python
unique_salary_df = player_salary_df.drop_duplicates("Player", keep = "last")
unique_salary_df
```

```python
# Save DataFrame to csv
unique_salary_df.to_csv("data/Unique_Salary_clean.csv", index = False)
```

## Average Salary and Performance

```python
average_player_salary_df = player_salary_df.groupby("Player").mean()
average_player_salary_df
```

```python
# Save DataFrame to csv
average_player_salary_df.to_csv("data/Average_Salary_clean.csv", index = False)
```
