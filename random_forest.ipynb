```python
from sklearn import tree
import pandas as pd
import os
```

```python
df = pd.read_csv(os.path.join("data", "final_dataset.csv"))
df.head()
```

```python
#data = df.drop(["Championship Appearance Y/N", "Player", "Team", "Pos", "Year", "Championship Wins", "RunnerUp Count"], axis=1)
data = df.drop(["Player", "Team", "Pos", "Year", "Championship Wins", "RunnerUp Count"], axis=1)
feature_names = data.columns
data
```

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=400)
rf.fit(X_train, y_train)
```

```python
rf_pred = rf.predict(X_test)
```

```python
rf_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, rf_pred)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

# Post 1980 - All complete stats from 1980-2017, 17291 rows 

```python
data_80_1 = df.drop(["Player", "Team", "Pos", "Championship Wins", "RunnerUp Count"], axis=1)
data_80_2 = data_80_1[data_80_1["Year"] >= 1980]
data_post80 = data_80_2.drop(["Year"], axis=1)
feature_names = data_post80.columns
data_post80
```

```python
X = data_post80.iloc[:, :-1].values
y = data_post80.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=400)
rf.fit(X_train, y_train)
```

```python
rf_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, rf_pred)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```
